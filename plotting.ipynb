```python
import pickle
import matplotlib.pyplot as plt
```

```python
# evaluate r2 score by month
test_months = [year*100 + month for year in range(1996, 2018) for month in range(1, 13)]
test_months = [str(t) for t in test_months]
```

## Plotting for NN3 results

```python
with open("outputs/nn3/test_r2s.pkl", "rb") as f:
    nn3_test_r2s = pickle.load(f)

with open("outputs/nn3_50epochs/test_r2s.pkl", "rb") as f:
    nn3_50_test_r2s = pickle.load(f)

with open("outputs/elasticnet/test_r2s.pkl", "rb") as f:
    elasticnet_test_r2s = pickle.load(f)

plt.figure(figsize=(27, 6))
plt.plot(test_months, nn3_test_r2s, label="NN3 R^2 on test")
plt.plot(test_months, nn3_50_test_r2s, label="NN3 (50 epochs) R^2 on test")
plt.plot(test_months, elasticnet_test_r2s, label="Elastic net R^2 on test")
plt.axhline(0, color='r', linestyle="--")
# plt.plot(test_months, validation_r2s, label="R^2 in validation years")
plt.xticks(test_months[::5], rotation='vertical', fontsize=14)
plt.xlabel('Date (YYYYMM)', fontsize=20)
plt.ylabel('R^2 Score', fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
